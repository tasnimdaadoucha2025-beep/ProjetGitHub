# 2 / Préparation des données

```python
# Valeur manquantes
df.isnull()
```

```python
# le nombre de valeurs maquantes 
df.isnull().sum()
```

```python
# Effacer les colonnes inutiles
df.drop(columns=['certificate'], inplace=True)
```

```python
df.columns = df.columns.str.strip()
```

```python
# Eliminer les colonnes inutiles
df = df.drop(columns=['ranking of movie'])
```

```python
df.columns
```

```python
# Remplacer les valeurs manquantes par la  moyenne
df['metascore'] = df['metascore'].fillna(df['metascore'].mean())
```

```python
# Nettoyer la colonne des symboles 
df['GROSS COLLECTION'] = df['GROSS COLLECTION'].str.replace('$', '', regex=False)
df['GROSS COLLECTION'] = df['GROSS COLLECTION'].str.replace('M', '', regex=False)
```

```python
# Convertir en numérique 
df['GROSS COLLECTION'] = pd.to_numeric(df['GROSS COLLECTION'], errors='coerce')
```

```python
# Remplacer les valeurs manquantes par la  médianne
df['GROSS COLLECTION'] = df['GROSS COLLECTION'].fillna(df['GROSS COLLECTION'].median())
```

```python
# Vérification 
df.isnull().sum()
```

```python
# les valeurs dupliquée
df.duplicated().sum()
```

```python
# Renommer les Variables 
df = df.rename(columns={
    'movie name': 'Title',
    'Year': 'Released Year',
    'runtime': 'Duration Min',
    'genre': 'Genre',
    'RATING': 'IMDB_Rating',
    'metascore': 'Meta Score',
    'DIRECTOR': 'Director',
    'ACTOR 1': 'Main Actor 1',
    'ACTOR 2': 'Main Actor 2',
    'ACTOR 3': 'Main Actor 3',
    'ACTOR 4': 'Main Actor 4',
    'votes': 'Num_Vote',
    'GROSS COLLECTION': 'Revenue',
    'DETAIL ABOUT MOVIE':'Movie_Details'
})
```

```python
df.columns
```

```python
# Nettoyer et convertir une colonne to  numeric 
df['Num_Vote'] = df['Num_Vote'].str.replace(',', '')
df['Num_Vote'] = pd.to_numeric(df['Num_Vote'], errors='coerce')
```

```python
df.dtypes
```

```python
# Convertir en numérique 
df['Released Year'] = (
    pd.to_numeric(df['Released Year'], errors='coerce')
    .abs()
    .astype('Int64')
)
```

```python
df.dtypes
```

```python
# Nettoyer la colonne puis  Convertir en numérique 
df['Duration Min'] = df['Duration Min'].str.replace('min', '')
df['Duration Min'] = pd.to_numeric(df['Duration Min'], errors='coerce')
```

```python
df.dtypes
```

# 4/ Analyse des KPI 

```python
#Films récents mieux notés ?
df.groupby('Released Year')['IMDB_Rating'].mean()
```

```python
#Décennie la plus rentable
df['Decade'] = (df['Released Year']//10)*10
df.groupby('Decade')['Revenue'].sum()
```

```python
#Moyenne des notes IMDB par genre
df.groupby('Genre')['IMDB_Rating'].mean()
```

```python
#Top 10 des genres par revenu total
df.groupby('Genre')['Revenue'].sum().sort_values(ascending=False).head(10)
```

```python
#Top 10 des réalisateurs par note moyenne
df.groupby('Director')['IMDB_Rating'].mean().sort_values(ascending=False).head(10)
```

```python
#Top 10 des acteurs principaux par revenu total
df.groupby('Main Actor 1')['Revenue'].sum().sort_values(ascending=False).head(10)
```

```python
df.groupby('Main Actor 2')['Revenue'].sum().sort_values(ascending=False).head(10)
```

```python
df.groupby('Main Actor 3')['Revenue'].sum().sort_values(ascending=False).head(10)
```

```python
df.groupby('Main Actor 4')['Revenue'].sum().sort_values(ascending=False).head(10)
```

```python
#Top 10 des genres ayant la plus longue durée moyenne
df.groupby('Genre')['Duration Min'].mean().sort_values(ascending=False).head(10)
```
